````python
import re
import networkx as nx

graph = """Node 1 is connected to nodes 0 (weight: 10), 2 (weight: 7), 10 (weight: 10), 9 (weight: 2), 5 (weight: 8), 7 (weight: 7), 8 (weight: 10), 3 (weight: 2).
Node 0 is connected to node 1 (weight: 10).
Node 2 is connected to node 1 (weight: 7).
Node 10 is connected to node 1 (weight: 10).
Node 9 is connected to node 1 (weight: 2).
Node 5 is connected to node 1 (weight: 8).
Node 7 is connected to node 1 (weight: 7).
Node 8 is connected to nodes 1 (weight: 10), 6 (weight: 1).
Node 3 is connected to node 1 (weight: 2).
Node 6 is connected to nodes 8 (weight: 1), 4 (weight: 10).
Node 4 is connected to node 6 (weight: 10)."""

# Regular expression to extract edges with weights
pattern = r'Node (\d+) is connected to nodes? ([\d, ()weight:]+)\.'
matches = re.findall(pattern, graph)

# Create an undirected graph
G = nx.Graph()

# Add edges with weights to the graph
for node, connections in matches:
    node = int(node)
    edges = re.findall(r'(\d+) \(weight: (\d+)\)', connections)
    for neighbor, weight in edges:
        G.add_edge(node, int(neighbor), weight=int(weight))

print(G.edges())
# Calculate the distance of the shortest path from node 10 to node 4
try:
    length = nx.shortest_path_length(G, source=10, target=4, weight='weight')
    print(length)
except:
    print(f"There is no path between node 10 and node 4")
````
